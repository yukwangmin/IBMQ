```python
b = np.pi / 4

#a = 0.179636 # 2 qubits
#a = 0.181178 # 3 qubits
a = 0.18169  # 10 & 14 qubits

theta = np.arcsin(np.sqrt(a))

print('true a: ', a)
print('theta: ', theta)
print('prob of theta: ', (np.sin(1*theta))**2)
print('prob of 3theta: ', (np.sin(3*theta))**2)
print('prob of 5theta: ', (np.sin(5*theta))**2)
print('prob of 9theta: ', (np.sin(9*theta))**2)
print('prob of 17theta: ', (np.sin(17*theta))**2)
print('prob of 33theta: ', (np.sin(33*theta))**2)
print('\n\n')


backend = BasicAer.get_backend('qasm_simulator')

# n: number of qubits of the problem domain
# m: the length of the MLQAE circuits, m+1 circuits.
# b: integral interval: [0, b]
# shots: 
# trial:
# backend:
# true_a: the exact value of the integral. used for the error estimation.
est_theta, est_a, rel_err_theta, rel_err_a = stat_mlae(10, 3, b, 1024, 4, backend, a)





print('est_theta:', est_theta)
print('est_a:', est_a)

print('Average of theta: ', stat.mean(est_theta))    
print('Average of a: ', stat.mean(est_a))
print('Average of relative error of theta: ', stat.mean(rel_err_theta))
print('Average of relative error of a: ', stat.mean(rel_err_a))

print('Stdev of theta: ', stat.stdev(est_theta))    
print('Stdev of a: ', stat.stdev(est_a))
print('Stdev of relative error of theta: ', stat.stdev(rel_err_theta))
print('Stdev of relative error of a: ', stat.stdev(rel_err_a))

print('Min of theta: ', min(est_theta))    
print('Min of a: ', min(est_a))
print('Min of relative error of theta: ', min(rel_err_theta))
print('Min of relative error of a: ', min(rel_err_a))

print('Max of theta: ', max(est_theta))    
print('Max of a: ', max(est_a))
print('Max of relative error of theta: ', max(rel_err_theta))
print('Max of relative error of a: ', max(rel_err_a))




```
